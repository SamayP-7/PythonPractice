```python
import matplotlib.pyplot as plt

x=['Rohan','Rahul','Suraj','Akshay']
y=[57,68,75,82]

# color  r
# marker x o ^
# pattern of line -
plt.plot(x,y,'bo-',markeredgecolor='red')
plt.xlabel('Students name')
plt.ylabel('Markes')
plt.title("Student Report")
plt.show()  # optinal 
```

```python
name=['Rohan','Rahul','Suraj','Akshay']
eng=[57,68,75,82]

sci=[88,61,72,79]
maths=[78,89,57,62]
bar1=[0,2,4,6]
bar2=[i+0.5 for i in bar1]
bar3=[i+0.5 for i in bar2]

plt.bar(bar1,eng,width=0.4,color="red")
plt.bar(bar2,sci,width=0.4)
plt.bar(bar3,maths,width=0.4)
plt.xlabel('Students name')
plt.ylabel('Markes')
plt.title("Student Report")
plt.xticks(bar2,name)
plt.legend(['eng','sci','maths'])
plt.show()  # optinal 
```

```python
#Boxplot
# describe --> min,max,25%,50% & 75%

marks=[60,62,64,65,68,62,64,65,98]

# empty circle is called outlier value
# which tells the value is not in range
plt.boxplot(marks)
plt.show()  # optinal 
```

```python
#histogram
# frq distribution graph

values=[5,7,8,10,12,22,28,33,39,41,43,56,57,62,63]

# 0-10  - 4
#10-20
#20-30
#40-50
#50-60
plt.hist(values,ec='red',bins=[0,15,30,45,60,75])
plt.show()  # optinal 

```

```python
# pie chart
expenses=['rent','maintenance','tea','other']
amt=[20000,3500,1200,4000]

plt.pie(amt,labels=expenses,autopct="%.2f%%")

plt.show()
```
