```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# titanic , iris
df=sns.load_dataset('iris')
print(df)

# relplot --> relation plot

sns.relplot(x='petal_length',y='sepal_width',data=df,hue='species',col='species',col_wrap=2)

plt.show()

```

```python
sns.pairplot(df)



plt.show()
```

```python
# histogram + scatter plot --> jointplot
sns.jointplot(x='sepal_length',y='species',data=df)
```

```python
sns.distplot(df.petal_length,rug=True)
```

```python
sns.kdeplot(df.sepal_width)
```

```python
sns.barplot(x='species',y='sepal_length',data=df)
```

```python
sns.rugplot(df.sepal_length)
```

```python
sns.countplot(x='sex',data=df)
```
