```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# titanic , iris
df=sns.load_dataset('titanic')
print(df)

# relplot --> relation plot

sns.relplot(x='age',y='who',data=df,hue='survived',style='sex',col='class',col_wrap=2)

plt.show()

```

```python
sns.pairplot(df)



plt.show()
```

```python
# histogram + scatter plot --> jointplot
sns.jointplot(x='age',y='class',data=df)
```

```python
sns.distplot(df.fare,rug=True)
```

```python
sns.kdeplot(df.fare)
```

```python
sns.barplot(x='class',y='fare',data=df)
```

```python
sns.rugplot(df.fare)
```

```python
sns.countplot(x='sex',data=df)
```
